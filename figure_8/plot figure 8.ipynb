```python
import matplotlib.pyplot as plt
import json
import numpy as np
from scipy.signal import savgol_filter
```

```python
with open('./ENERGIES.json', 'r') as f:
    ENERGIES = json.load(f)
with open('./ENERGIES_RAND.json', 'r') as f:
    ENERGIES_RAND = json.load(f)
with open('./ENERGIES_RAND2.json', 'r') as f:
    ENERGIES_RAND2 = json.load(f)
with open('./RANKS.json', 'r') as f:
    RANKS = json.load(f)
```

```python
ENERGIES = np.array(ENERGIES)
ENERGIES_RAND = np.array( ENERGIES_RAND)
ENERGIES_RAND2 = np.array( ENERGIES_RAND2)
RANKS = np.array(RANKS)
```

```python
size = len(ENERGIES)
size
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(211)
plt.plot( RANKS, linewidth = 5, color = '#F80EF8')
plt.grid()
plt.xlim((0,1494))
plt.xlabel(r'$N_D$', fontsize=20)
plt.ylabel(r'$rank[W_S]$', fontsize = 20)

ax2 = fig.add_subplot(212)
plt.plot( np.linspace( 1494-size, 1494, size), 
         savgol_filter( np.log( ENERGIES_RAND2[:-1] ), 35, 2 ), linewidth = 5, color = '#C81E55', label = 'random' )
plt.plot( np.linspace( 1494-size, 1494, size), 
         np.log( ENERGIES ), linewidth = 5, color = '#F80EF8', label = 'optimized' )
plt.grid()
plt.legend(loc=3)
plt.xlim((1494-size, 1494))
plt.xlabel(r'$N_D$', fontsize = 20)
plt.ylabel(r'$log(\mathcal{E})$', fontsize = 20)
```
